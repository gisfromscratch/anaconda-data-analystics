```python
import pandas as pd
```

```python
header = ['geonameid','name','asciiname','alternatenames','latitude','longitude','feature class','feature code','country code','cc2','admin1 code','admin2 code','admin3 code','admin4 code','population','elevation','dem','timezone','modification date']
geonames = pd.read_csv('D:/Geonames/allCountries.txt', delimiter='\t', header=None, names=header, low_memory=False)
```

```python
geonames.head()
```

```python
dates = set(geonames['modification date'])
print('%d unique dates' % len(dates))
featureCodes = set(geonames['feature code'])
print('%d unique feature codes' % len(featureCodes))
countryCodes = set(geonames['country code'])
print('%d unique country codes' % len(countryCodes))
```

```python
modifications = pd.read_csv('D:/Geonames/modifications-2018-02-24.txt', delimiter='\t', header=None, names=header, low_memory=False)
modifications.head()
```

```python
geonamesModifications = pd.merge(geonames, modifications, on='geonameid', how='inner')
geonamesModifications.head()
```
