```python
import pandas as pd
```

```python
header = ['geonameid','name','asciiname','alternatenames','latitude','longitude','feature class','feature code','country code','cc2','admin1 code','admin2 code','admin3 code','admin4 code','population','elevation','dem','timezone','modification date']
geonames = pd.read_csv('D:/Geonames/cities1000.txt', delimiter='\t', header=None, names=header, low_memory=False)
```

```python
geonames.head()
```

```python
dates = set(geonames['modification date'])
print('%d unique dates' % len(dates))
featureCodes = set(geonames['feature code'])
print('%d unique feature codes' % len(featureCodes))
countryCodes = set(geonames['country code'])
print('%d unique country codes' % len(countryCodes))
```

```python
modifications = pd.read_csv('D:/Geonames/modifications-2018-02-24.txt', delimiter='\t', header=None, names=header, low_memory=False)
modifications.head()
```

```python
geonamesModifications = pd.merge(geonames, modifications, on='geonameid', how='inner')
geonamesModifications.head()
```

```python
modificationsConcat = pd.concat([geonames.set_index('geonameid'), modifications.set_index('geonameid')], axis='columns', keys=['Data', 'Mods'], join='inner')
modificationsSwap = modificationsConcat.swaplevel(axis='columns')[geonames.columns[1:]]
```

```python
import numpy as np

def highlight_diff(data, color='yellow'):
    attr = 'background-color: %s' % color
    other = data.xs('Data', axis='columns', level=-1)
    return pd.DataFrame(np.where(data.ne(other, level=0), attr, ''), index=data.index, columns=data.columns)

def print_diff(data):
    return data[0] if data[0] == data[1] else '{} ---> {}'.format(*data)

styledDiff=modificationsSwap.style.apply(highlight_diff, axis=None)
styledDiff

#diffPanel = pd.Panel(dict(Data=modificationsConcat['Data'], Mods=modificationsConcat['Mods']))
#diffResult = diffPanel.apply(print_diff, axis=0)
#diffResult
```
